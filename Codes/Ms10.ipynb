```
import pandas as pd
import random
import uuid
from faker import Faker
from datetime import datetime
from datetime import timedelta

from decimal import *
getcontext().prec = 4

class SalesLT:
    def generateAddressData(self,count): 

        Address ={} 
        addressid = 1
        for i in range(0, count): 
            Address[i]={} 
            Address[i]['AddressId']= addressid
            Address[i]['AddressLine1']= fake.street_address()
            Address[i]['AddressLine2']= fake.street_name()
            Address[i]['City']= fake.city() 
            Address[i]['StateProvince']= fake.state() 
            Address[i]['CountryRegion']= fake.country()
            Address[i]['PostalCode']= fake.postcode()
            Address[i]['rowguid']= str(uuid.uuid4())
            Address[i]['ModifiedDate']= datetime.strptime(fake.date()+" "+fake.time(), '%Y-%m-%d %H:%M:%S')
            addressid = addressid + 1

        return Address

    #----------------------------------------------------------------------------------------------------------------------------------------------#
    
    def generateCustomerData(self,count): 

        Customer ={} 
        customerid = 1000
        for i in range(0, count): 
            Customer[i]={} 
            Customer[i]['CustomerId']= customerid
            Customer[i]['NameStyle']= fake.pybool()
            Customer[i]['Title']= fake.prefix()
            Customer[i]['FirstName']= fake.first_name() 
            Customer[i]['MiddleName']= ''
            Customer[i]['LastName']= fake.last_name()
            Customer[i]['Suffix']= fake.suffix()
            Customer[i]['CompanyName']= fake.company()
            Customer[i]['SalesPerson']= "adventure-works\\"+fake.first_name()
            Customer[i]['EmailAddress']= fake.email()
            Customer[i]['Phone']= fake.phone_number()
            Customer[i]['PasswordHash']= fake.md5()
            Customer[i]['PasswordSalt']= fake.password()
            Customer[i]['rowguid']= str(uuid.uuid4())
            Customer[i]['ModifiedDate']= datetime.strptime(fake.date()+" "+fake.time(), '%Y-%m-%d %H:%M:%S')
            customerid = customerid + 1

        return Customer

    #----------------------------------------------------------------------------------------------------------------------------------------------#
    
    def generateCustomerAddressData(self,count): 

        CustomerAddress ={}
        addressid = 1
        customerid = 1000
        for i in range(0, count): 
            CustomerAddress[i]={} 
            CustomerAddress[i]['CustomerId']= customerid
            CustomerAddress[i]['AddressId']= addressid
            CustomerAddress[i]['AddressType']= fake.random_element(elements=('Main Office', 'Shipping'))
            CustomerAddress[i]['rowguid']= str(uuid.uuid4())
            CustomerAddress[i]['ModifiedDate']= datetime.strptime(fake.date()+" "+fake.time(), '%Y-%m-%d %H:%M:%S')
            addressid = addressid + 1
            customerid = customerid + 1

        return CustomerAddress

    #----------------------------------------------------------------------------------------------------------------------------------------------#
    
    def generateProductData(self,count): 

        Product ={} 
        productid = 2000
        for i in range(0, count): 
            Product[i]={} 
            Product[i]['ProductId']= productid
            Product[i]['Name']= fake.random_element(elements=('HL Mountain Frame', 'LL Road Frame', 'Mountain Bike Socks', 'ML Fork','Racing Socks'))
            Product[i]['ProductNumber']= fake.bothify(text='??-?###-?', letters='ABCDE')
            Product[i]['Color']= fake.safe_color_name() 
            Product[i]['StandardCost']= fake.pydecimal(3,4,True,1,100)
            Product[i]['ListPrice']= fake.pydecimal(3,4,True,1,100)
            Product[i]['Size']= fake.random_element(elements=('S','M','L','XL'))
            Product[i]['Weight']= fake.pydecimal(2,2,True,33,66)
            Product[i]['Name']= Product[i]['Name'] + " - " + Product[i]['Color'] + "," + Product[i]['Size']
            Product[i]['ProductCategoryId']= fake.random_int(min=9000,max=9100)
            Product[i]['ProductModelId']= fake.random_int(min=4000,max=4150)
            Product[i]['ModifiedDate']= datetime.strptime(fake.date()+" "+fake.time(), '%Y-%m-%d %H:%M:%S')
            Product[i]['SellStartDate']= Product[i]['ModifiedDate']+timedelta(days=-10)
            Product[i]['SellEndDate']= Product[i]['ModifiedDate']+timedelta(days=-5)
            Product[i]['DiscontinuedDate']= Product[i]['SellEndDate']+timedelta(days=1)
            Product[i]['ThumbNailPhoto']= fake.binary(length=64)
            Product[i]['ThumbnailPhotoFileName']= fake.name() + ".gif"
            Product[i]['rowguid']= str(uuid.uuid4())

            productid = productid + 1

        return Product

    #----------------------------------------------------------------------------------------------------------------------------------------------#
        
    def generateProductCategoryData(self,count): 

        ProductCategory={} 
        productcategoryid = 9000

        for i in range(0, count): 
            ProductCategory[i]={} 
            ProductCategory[i]['ProductCategoryID']= productcategoryid
            ProductCategory[i]['ParentProductCategoryID']= productcategoryid
            ProductCategory[i]['Name']= fake.word() 
            ProductCategory[i]['rowguid']= str(uuid.uuid4())
            ProductCategory[i]['ModifiedDate']= datetime.strptime(fake.date()+" "+fake.time(), '%Y-%m-%d %H:%M:%S')

            productcategoryid = productcategoryid + 1


        return ProductCategory

    #----------------------------------------------------------------------------------------------------------------------------------------------#
    
    def generateProductDescriptionData(self,count): 

        ProductDescription={} 
        productdescriptionid = 5000

        for i in range(0, count): 
            ProductDescription[i]={} 
            ProductDescription[i]['ProductDescriptionID']= productdescriptionid
            ProductDescription[i]['Description']= fake.text() 
            ProductDescription[i]['rowguid']= str(uuid.uuid4())
            ProductDescription[i]['ModifiedDate']= datetime.strptime(fake.date()+" "+fake.time(), '%Y-%m-%d %H:%M:%S')

            productdescriptionid = productdescriptionid + 1


        return ProductDescription

    #----------------------------------------------------------------------------------------------------------------------------------------------#

    def generateProductModelData(self,count): 

        ProductModel={} 
        productmodelid = 4000

        for i in range(0, count): 
            ProductModel[i]={} 
            ProductModel[i]['ProductModelID']= productmodelid
            ProductModel[i]['Name']= fake.random_element(elements =('HL Mountain Frame', 'LL Road Frame', 'Mountain Bike Socks', 'ML Fork','Racing Socks'))
            ProductModel[i]['CatalogDescription']= fake.text() 
            ProductModel[i]['rowguid']= str(uuid.uuid4())
            ProductModel[i]['ModifiedDate']= datetime.strptime(fake.date()+" "+fake.time(), '%Y-%m-%d %H:%M:%S')

            productmodelid = productmodelid + 1


        return ProductModel

    #----------------------------------------------------------------------------------------------------------------------------------------------#

    def generateProductModelProductDescriptionData(self,count): 

        ProductModelProductDescription={} 
        productmodelid = 4000
        productdecriptionid = 5000
        culture = "en"
        for i in range(0, count): 
            ProductModelProductDescription[i]={} 
            ProductModelProductDescription[i]['ProductModelID']= productmodelid
            ProductModelProductDescription[i]['ProductDescriptionID']= productdecriptionid
            ProductModelProductDescription[i]['Culture']= culture
            ProductModelProductDescription[i]['rowguid']= str(uuid.uuid4())
            ProductModelProductDescription[i]['ModifiedDate']= datetime.strptime(fake.date()+" "+fake.time(), '%Y-%m-%d %H:%M:%S')
            if i % 6 == 0:
                productmodelid = productmodelid + 1
                culture = fake.random_element(elements = ('en','ar','fr','th','he','zh-cht'))
            productdecriptionid = productdecriptionid + 1

        return ProductModelProductDescription

    #----------------------------------------------------------------------------------------------------------------------------------------------#
    
    def generateSalesOrderDetailData(self,count): 

        SalesOrderDetail ={} 
        salesorderid = 3000
        salesorderdetailid = 5000
        productid = 2000
        for i in range(0, count): 
            SalesOrderDetail[i]={} 
            SalesOrderDetail[i]['SalesOrderID']= salesorderid
            SalesOrderDetail[i]['SalesOrderDetailID']= salesorderdetailid
            SalesOrderDetail[i]['OrderQty']= fake.random_int(0, 25)
            SalesOrderDetail[i]['ProductID']= productid
            SalesOrderDetail[i]['UnitPrice']= fake.pydecimal(4,4,True,30,1500)
            SalesOrderDetail[i]['UnitPriceDiscount']= fake.pydecimal(1,2,False,0,1)
            SalesOrderDetail[i]['LineTotal']= SalesOrderDetail[i]['OrderQty'] * (SalesOrderDetail[i]['UnitPrice'] - SalesOrderDetail[i]['UnitPriceDiscount'])
            SalesOrderDetail[i]['rowguid'] = str(uuid.uuid4())
            SalesOrderDetail[i]['ModifiedDate'] = datetime.strptime(fake.date()+" "+fake.time(), '%Y-%m-%d %H:%M:%S')
            productid = productid + 1
            salesorderid =salesorderid + 1
            salesorderdetailid = salesorderdetailid + 1

        return SalesOrderDetail

    #----------------------------------------------------------------------------------------------------------------------------------------------#

    def generateSalesOrderHeaderData(self,count): 

        SalesOrderHeader ={} 
        salesorderid = 3000
        customerid = 1000
        for i in range(0, count): 
            SalesOrderHeader[i]={} 
            SalesOrderHeader[i]['SalesOrderID']= salesorderid
            SalesOrderHeader[i]['RevisionNumber']= fake.random_int(1,5)
            SalesOrderHeader[i]['ModifiedDate']= datetime.strptime(fake.date()+" "+fake.time(), '%Y-%m-%d %H:%M:%S')
            SalesOrderHeader[i]['OrderDate']= SalesOrderHeader[i]['ModifiedDate']+timedelta(days=-10)
            SalesOrderHeader[i]['DueDate']= SalesOrderHeader[i]['ModifiedDate']
            SalesOrderHeader[i]['ShipDate']= SalesOrderHeader[i]['ModifiedDate']+timedelta(days=-7)
            SalesOrderHeader[i]['Status']= 5
            SalesOrderHeader[i]['OnlineOrderFlag']= fake.pybool()
            SalesOrderHeader[i]['SalesOrderNumber']= "SO"+ str(SalesOrderHeader[i]['SalesOrderID'])
            SalesOrderHeader[i]['PurchaseOrderNumber']= fake.bothify(text= 'PO#########')
            SalesOrderHeader[i]['AccountNumber'] = fake.bothify('10-4020-000###')
            SalesOrderHeader[i]['CustomerID']= customerid
            SalesOrderHeader[i]['ShipToAddressID']= fake.random_int(620,1150)
            SalesOrderHeader[i]['BillToAddressID']= SalesOrderHeader[i]['ShipToAddressID']
            SalesOrderHeader[i]['ShipMethod']= "CARGO TRANSPORT 5"
            SalesOrderHeader[i]['CreditCardApprovalCode']= ""
            SalesOrderHeader[i]['SubTotal']= fake.pydecimal(3,4,True,100,300)
            SalesOrderHeader[i]['TaxAmt']= SalesOrderHeader[i]['SubTotal'] * Decimal(0.08000003)
            SalesOrderHeader[i]['Freight']= SalesOrderHeader[i]['TaxAmt'] * Decimal(0.31249973)
            SalesOrderHeader[i]['TotalDue']= SalesOrderHeader[i]['SubTotal'] + SalesOrderHeader[i]['TaxAmt'] + SalesOrderHeader[i]['Freight'] 
            SalesOrderHeader[i]['Comment']= fake.text()
            SalesOrderHeader[i]['rowguid']= str(uuid.uuid4())

            salesorderid = salesorderid + 1
            customerid = customerid + 1

        return SalesOrderHeader
```

```
def readingConfigFile(filepath):
    """ This function reads the configuration file from ADLS """
    try:
        file = urlopen(filepath)
    except HTTPError:
        print("Incorrect URL.")
        return
    data = ""
    for line in file: data += line.decode()
    try:
        config = json.loads(data)
        return config
    except ValueError as e:
        return "Decoding JSON file failed"
```
